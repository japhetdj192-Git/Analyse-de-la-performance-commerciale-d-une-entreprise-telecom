```python
# --- 1. DÉTECTION DES ANOMALIES (Nettoyage) ---
# On utilise la méthode de l'Écart Interquartile (IQR) pour isoler les montants aberrants
def detecter_anomalies(df):
    Q1 = df['Chiffre_Affaires_XAF'].quantile(0.25)
    Q3 = df['Chiffre_Affaires_XAF'].quantile(0.75)
    IQR = Q3 - Q1
    seuil_haut = Q3 + 1.5 * IQR

    anomalies = df[df['Chiffre_Affaires_XAF'] > seuil_haut]
    print(f"⚠️ Anomalies détectées : {len(anomalies)} transactions au-dessus de {seuil_haut:,.0f} XAF.")
    return anomalies, seuil_haut
```

```python
# Verification format date
df['Date'] = pd.to_datetime(df['Date'])
# let's get the months names
df['Months']=df['Date'].dt.strftime('%b')
```

```python
# Nettoyage : On retire les anomalies et les valeurs nulles pour l'analyse
anomalies_df, seuil = detecter_anomalies(df)
df_clean = df[(df['Chiffre_Affaires_XAF'] > 0) & (df['Chiffre_Affaires_XAF'] <= seuil)].copy()

# --- 2. ANALYSE DE MASSE (Statistiques descriptives) ---
print("\n--- STATISTIQUES PAR VILLE ---")
print(df_clean.groupby('Ville')['Chiffre_Affaires_XAF'].describe().round(0))
```

# 3. Identifier les clients à risque de désabonnement (churn)

```python
# date de la dernière transaction
last_transaction_date = df_clean.groupby('Customer_ID')['Date'].max().reset_index()
last_transaction_date.rename(columns={'Date': 'Last_Transaction_Date'}, inplace=True)
# date la plus récente
current_data_date = df_clean['Date'].max()
# calculer le nombre de jours depuis la dernière transaction
last_transaction_date['Days_Since_Last_Transaction'] = (current_data_date - last_transaction_date['Last_Transaction_Date']).dt.days
# seuil pour le risque de churn
churn_risk_threshold_days = 60
churn_risk_customers = last_transaction_date[last_transaction_date['Days_Since_Last_Transaction'] > churn_risk_threshold_days]
if not churn_risk_customers.empty:
    print(f"Clients à risque de désabonnement (pas d'activité depuis > {churn_risk_threshold_days} jours):\n")
    print(churn_risk_customers.sort_values(by='Days_Since_Last_Transaction', ascending=False).head(10))
else:
    print(f"Aucun client identifié comme à risque de désabonnement basé sur un seuil de {churn_risk_threshold_days} jours.")

import matplotlib.pyplot as plt
import seaborn as sns

# visualisation
plt.figure(figsize=(7, 3))
sns.histplot(last_transaction_date['Days_Since_Last_Transaction'], bins=50, kde=True)
plt.axvline(churn_risk_threshold_days, color='red', linestyle='--', label=f'Seuil de risque de churn ({churn_risk_threshold_days} jours)')
plt.title('Distribution des Jours depuis la Dernière Transaction par Client')
plt.xlabel('Jours depuis la Dernière Transaction')
plt.ylabel('Nombre de Clients')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

**Estimation du Retour sur Investissement (ROI) des Actions Commerciales :**

L'estimation précise du ROI nécessiterait des données supplémentaires et des expérimentations (A/B testing) pour chaque action. Cependant, une estimation qualitative peut être faite :

*   **Actions de Réactivation (Churn) :** Le ROI est potentiellement très élevé. Acquérir un nouveau client coûte généralement 5 à 25 fois plus cher que de retenir un client existant. Réduire le churn de 5% peut augmenter les profits de 25% à 95% (selon Bain & Company). Un investissement modéré dans des campagnes de réactivation pourrait donc générer un retour significatif.

*   **Actions Ciblées sur les Villes et Produits Sous-Performants :** Le ROI dépendra de l'ampleur de l'investissement marketing et des améliorations de produit/service. En ciblant des marchés spécifiques, les efforts marketing peuvent être plus efficaces et donc potentiellement générer un ROI positif en augmentant la part de marché et le chiffre d'affaires dans ces segments.

*   **Programme de Fidélité (Clients à Forte Valeur) :** Le ROI est généralement positif à long terme. Les clients fidèles dépensent plus et sont plus susceptibles de recommander l'entreprise. Bien que l'investissement initial puisse être important, l'augmentation de la valeur vie client (CLTV) et la réduction des coûts d'acquisition compensent largement.

*   **Analyse Post-Mortem :** Le ROI de cette action est indirect mais crucial. En identifiant la cause d'une baisse, l'entreprise peut éviter de futures pertes substantielles, ce qui représente une économie de revenus potentiellement très élevée.
